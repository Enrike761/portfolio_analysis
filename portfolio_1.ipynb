```python
df.fillna(method='ffill', inplace=True) #rellenamos los Nan con los valores prÃ³ximos para poder calcular los retornos porcentuales diarios
```

```python
df.isna().sum()
```

```python
df
```

```python
df.set_index('date', inplace=True) #convertimos date en Ã­ndice para calcular los porcentajes
```

```python
returns = df.pct_change() #Calcula la variaciÃ³n porcentual entre los valores de cada fila y la anterior. El resultado serÃ¡ un DataFrame con los retornos porcentuales diarios.
```

```python
returns
```

```python
#en la primera columna son Nans al no poder calcular el porcentaje por no tener comparativa
```

```python
correlation_matrix = returns.corr()
```

```python
correlation_matrix
```

```python
#podemos observar una correlaciÃ³n neutra en Asset1 respecto al resto (primera columna), es decir, los distintos activos siguen una direcciÃ³n random ya que la correlaciÃ³n es cercana al 0, siendo 1 una direcciÃ³n homogÃ©nea perfecta y -1 opuesta
```

```python
plt.scatter(returns["Asset1"], returns["Asset2"], color='blue')
```

```python
plt.scatter(returns["Asset3"], returns["Asset5"], color='green')
```
